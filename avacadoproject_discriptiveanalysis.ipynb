```python
m=data["AveragePrice"].mean()
```

```python
#mean average price?

print("The average price is ", round(m,2)) #  round is built in function
```

```python
#what is standard deviation of average price?
s=data["AveragePrice"].std()
print(" The standard deviation in average price is:" ,round(s,2))
```

```python
# what are the unique type values?
data["type"].unique()
```

```python
#value counts for each data type
data["type"].value_counts()
```

```python
# groupby type of data and describe them.

data.groupby("type").describe()
```

```python
#decribe descriptive statistic for average column groupby type?
data_group_type=data.groupby("type")
data_group_type["AveragePrice"].describe()
```

```python
#calculate average price between the different type?#

data["AveragePrice"].groupby(data["type"]).mean()
```

```python
#show distribution of average prices using histogram?
data.hist(column="AveragePrice", bins =30, figsize =(4,4))
plt.xlabel("price")
plt.ylabel("count")
plt.title("distribution of avacado average price")
plt.show()
```

```python
#using seasborn libraryn to create distribution plot of "avarage price"?
plt.figure(figsize=(5,5))
ax=sns.displot(data["AveragePrice"],color="b")
plt.title("distribution of avacado average price")
plt.show()
```

```python
#overlay the distribution of average price grouped by type of data?
plt.figure(figsize=(15,5))
sns.histplot(data["AveragePrice"][data["type"]=="conventional"] ,color = "r", label='conventional')
sns.histplot(data["AveragePrice"][data["type"]=="organic"], color="b",label="organic")
plt.title("average price by type")
```

```python
# make boxplot using boxplot comparing  average price by type.
data.boxplot(column="AveragePrice", by ="type", figsize=(5,5))
plt.show()
```

```python
#Using boxplot using seaborn 
plt.figure(figsize=(12,5))
sns.boxplot( y= "type", x= "AveragePrice",data=data)
plt.xlim([0,4])
plt.show
```
